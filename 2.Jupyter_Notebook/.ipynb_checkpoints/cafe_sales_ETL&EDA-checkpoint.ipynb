```python
# Create a deep copy of the original data
cafe_cleaned = cafe.copy()

print("Original Data Shape:", cafe.shape)
print("Cleaning Copy Created!")
```

```python
import numpy as np
# 1. Standardize all garbage strings to NaN so Pandas can "see" them
garbage = ['ERROR', 'UNKNOWN', '', ' ']
cafe_cleaned.replace(garbage, np.nan, inplace=True)

# 2. Drop rows where Transaction_Date is missing (The 4.6%)
cafe_cleaned.dropna(subset=['Transaction_Date'], inplace=True)

print("New Data Shape:", cafe_cleaned.shape)
```

```python
print("Unique Items:", cafe_cleaned['Transaction_ID'].unique())
```

```python
print(cafe_cleaned['Transaction_ID'].value_counts(dropna=False))
```

```python
print("Unique Items:", cafe_cleaned['Payment_Method'].unique())
```

```python
print(cafe_cleaned['Payment_Method'].value_counts(dropna=False))
```

```python
print("Unique Items:", cafe_cleaned['Location'].unique())
```

```python
print(cafe_cleaned['Location'].value_counts(dropna=False))
```

```python
#Fill Categorical gaps with "Not Recorded"
cat_cols = ['Item', 'Payment_Method', 'Location']
cafe_cleaned[cat_cols] = cafe_cleaned[cat_cols].fillna('Not Recorded')
print(cafe_cleaned[cat_cols].isnull().sum())
```

```python
print("Unique Items:", cafe_cleaned['Location'].unique())
```

```python
print("New Data Shape:", cafe_cleaned.shape)
```

```python
cafe_cleaned.head()
```

```python
cafe_cleaned.tail()
```

```python
#NUMARICAL 

#  Convert the column to numbers first 
cafe_cleaned['Quantity'] = pd.to_numeric(cafe_cleaned['Quantity'], errors='coerce')

# 1)Fill Quantity with the Median (the middle-ground guess)
cafe_cleaned['Quantity'] = cafe_cleaned['Quantity'].fillna(cafe_cleaned['Quantity'].median())

print("Quantity Median:", cafe_cleaned['Quantity'].median())
```

```python
print(cafe_cleaned['Quantity'].unique())
```

```python
cafe_cleaned.head()
```

```python
#  Convert Price to numbers first so Pandas can do math
cafe_cleaned['Price_Per_Unit'] = pd.to_numeric(cafe_cleaned['Price_Per_Unit'], errors='coerce')

# 2. Fill Price_Per_Unit with the Mode of that specific Item
# This looks at all "Coffees" to find the most common "Coffee" price
cafe_cleaned['Price_Per_Unit'] = cafe_cleaned.groupby('Item')['Price_Per_Unit'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
# This creates a table showing the most frequent price for each item
item_modes = cafe_cleaned.groupby('Item')['Price_Per_Unit'].agg(lambda x: x.mode()[0])

print("--- The Calculated Mode Prices per Item ---")
print(item_modes)
```

```python
# Convert Total_Spent to numeric so Python sees it as a math column
cafe_cleaned['Total_Spent'] = pd.to_numeric(cafe_cleaned['Total_Spent'], errors='coerce')

# 3. Recalculate using our clean Quantity and Price
# This fixes all NaNs and any original calculation errors
cafe_cleaned['Total_Spent'] = cafe_cleaned['Quantity'] * cafe_cleaned['Price_Per_Unit']
```

```python
print(cafe_cleaned.isnull().sum())
```

```python
cafe_cleaned.head()
```

```python
# Convert Quantity from 2.0 to 2
cafe_cleaned['Quantity'] = cafe_cleaned['Quantity'].astype(int)
```

```python
cafe_cleaned.head()
```

```python
#Convert Transaction_Date to a real Date format
cafe_cleaned['Transaction_Date'] = pd.to_datetime(cafe_cleaned['Transaction_Date'])

# Final Check to see the beautiful clean types
print("--- Finalized Data Types ---")
print(cafe_cleaned.dtypes)
```

```python
# 1. Final shape of clean data
print("Final shape:", cafe_cleaned.shape)

# 2. Check no nulls remain
print("\nNull values:\n", cafe_cleaned.isnull().sum())

# 3. Check data types are correct
print("\nData types:\n", cafe_cleaned.dtypes)

# 4. Check sample of clean data
cafe_cleaned.head()
```

```python
#count how many duplicate rows exist
duplicate_count = cafe_cleaned.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")
```

```python
# Flip the table so columns become rows, then check for duplicates
duplicate_cols = cafe_cleaned.T.duplicated()

print("--- Duplicate Column Check ---")
print(duplicate_cols)
print(duplicate_cols.shape)
```

```python
# 1. Filter out 'Not Recorded' items to ensure our analysis is pristine
clean_item_df = df_eda[df_eda['Item'] != 'Not Recorded']

# 2. Setup the canvas
plt.figure(figsize=(10, 6))

# 3. Create a barplot showing the average Total_Spent for each Item
# Seaborn automatically calculates the mean (average) for the Y-axis
sns.barplot(data=clean_item_df, x='Item', y='Total_Spent', hue='Item',palette='Set2', legend=False,errorbar=None) # Turns off the little uncertainty lines to keep it clean

# 4. Add labels and titles
plt.title('Average Order Bill Size by Menu Item', fontsize=14, fontweight='bold')
plt.xlabel('Menu Item Ordered', fontsize=12)
plt.ylabel('Average Total Spent (Rupees)', fontsize=12)
plt.tight_layout()

# 5. Show the plot
plt.show()
```

```python
# 1. Setup the canvas
plt.figure(figsize=(10, 6))

# 2. Create a barplot showing the average QUANTITY bought for each Item
sns.barplot(data=clean_item_df, x='Item', y='Quantity', hue='Item',palette='Set2', legend=False,errorbar=None)

# 3. Add labels and titles
plt.title('Average Units Purchased per Order by Menu Item', fontsize=14, fontweight='bold')
plt.xlabel('Menu Item Ordered', fontsize=12)
plt.ylabel('Average Quantity (Units)', fontsize=12)
plt.tight_layout()

# 4. Show the plot
plt.show()
```

```python
clean_item_df.groupby('Item')[['Quantity']].agg(['mean'])
```

```python
# Check revenue contribution per item

revenue_analysis = clean_item_df.groupby('Item').agg(
    Total_Revenue=('Total_Spent', 'sum'),
    Avg_Price=('Price_Per_Unit', 'mean'),
    Total_Orders=('Transaction_ID', 'count')
).sort_values('Total_Revenue', ascending=False)  #if we have multiple columns to sort values , then we have to give specific column name inside sort_values() OR if we have single column then we don't need to give column name inside sort_values()

print(revenue_analysis)


# Calculate revenue percentage per item
revenue_analysis['Revenue_%'] = (
    revenue_analysis['Total_Revenue'] / 
    revenue_analysis['Total_Revenue'].sum() * 100
).round(2)

print(revenue_analysis)
```

```python
#Where do Coffee and Juice fall in this strategy?
#They are my bestsellers by volume —
#But are they also strong revenue contributors?

```

```python
print(revenue_analysis.loc[['Coffee', 'Juice']])
```

```python
#Coffee contributes 8.90% of total revenue —
#It sits between Low Risk zone (Cookie 4%, Tea 6%) and High Risk zone (Salad 21%)
#So Coffee falls in a Medium Risk zone!

```

```python
# Quick check — how sensitive are Coffee buyers?
# Compare Coffee order count vs other items
print(revenue_analysis.sort_values('Revenue_%', ascending=False))
```

```python
# 1. Get the exact count and percentage of orders for each quantity
quantity_counts = df_eda['Quantity'].value_counts().sort_index()
quantity_percentages = (df_eda['Quantity'].value_counts(normalize=True).sort_index() * 100).round(2)

# 2. Combine them into a clean, readable summary table
quantity_summary = pd.DataFrame({
    'Number of Orders (Count)': quantity_counts,
    'Percentage of Total Orders (%)': quantity_percentages
})


print("--- Quantity Distribution Summary ---")
print(quantity_summary)

# 3. Create a quick visual countplot to see it clearly
plt.figure(figsize=(8, 5))
sns.countplot(data=df_eda, x='Quantity', palette='viridis', hue='Quantity', legend=False)
plt.title('How Many Orders Are Placed per Quantity Bucket?', fontsize=14, fontweight='bold')
plt.xlabel('Quantity of Items in a Single Bill', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#question 4 : Do customers prefer premium items naturally?
```

```python
# Calculate the percentage of total orders each item represents
revenue_analysis['Order_%'] = (
    revenue_analysis['Total_Orders'] / revenue_analysis['Total_Orders'].sum() * 100
).round(2)

# Drop the revenue-related columns
revenue_analysis = revenue_analysis.drop(columns=['Total_Revenue', 'Revenue_%'])

# Display the clean, volume-focused table
print(revenue_analysis.sort_values(by='Order_%', ascending=False))
```
